```python
# Query to get the total number of flights per year
query = """
SELECT *
FROM analytics.aircraft_detail
"""

# Execute the query and fetch the results into a DataFrame
aircraft_df = pd.read_sql(query, engine)


aircraft_df.head()
```

```python
def find_aircraft_dev_id(regist):
    df = aircraft_df.query(f"aircraft_registration == '{regist}'")
    if df.empty:
        raise ValueError(f"Aircraft with registration {regist} not found.")
    return df.iloc[0]["dev_id"] if not df.empty else "ID Not Found"

find_aircraft_dev_id("PK-HNA")
```

```python


agg_df = (
    raw_logbook_df.groupby("formatted_serial_number")
    .agg(
        flight_date=("date", "min"),
        hobbs_start=("start", "min"),
        hobbs_end=("end", "max"),
        total_flight_hours_decimal=("hours", "sum"),
        total_legs=("landings", "sum"),
    )
    .reset_index()
)

# Assign aircraft_id and raw_serial_number
agg_df["aircraft_registration"] = agg_df["formatted_serial_number"].str.split(
    "_", expand=True
)[1]
agg_df["raw_serial_number"] = agg_df["formatted_serial_number"].str.split(
    "_", expand=True
)[2]

logbook_df_columns = [
    "id",
    "doc_img",
    "flight_date",
    "hobbs_start",
    "hobbs_end",
    "total_flight_hours_decimal",
    "legs_start",
    "legs_end",
    "total_legs",
    "created",
    "updated",
    "is_verified",
    "status",
    # "aircraft_id",
    "created_by_id",
    "updated_by_id",
    "verified_by_id",
    "raw_serial_number",
    "formatted_serial_number",
]

# Create the logbook_df with the required columns
logbook_df = pd.DataFrame(columns=logbook_df_columns)
logbook_df = pd.concat([logbook_df, agg_df], ignore_index=True, sort=False)

logbook_df["total_flight_hours_decimal"] = (
    logbook_df["total_flight_hours_decimal"].astype(float).round(3)
)
# Map aircraft registration to dev_id using the find_aircraft_dev_id function

logbook_df = logbook_df.merge(
    aircraft_df[["aircraft_registration", "dev_id"]],
    how="left",
    on="aircraft_registration",
).rename({"dev_id": "aircraft_id"}, axis=1)

logbook_df
```

```python
import pymysql.cursors


def get_db_data(cursor : pymysql.cursors.Cursor):
    cursor.execute(get_data_query)
    return cursor.fetchall()

def filter_new_records(cursor: pymysql.cursors.Cursor, new_data_df: pd.DataFrame):
    """
    Filter out records that already exist in the database based on formatted_serial_number only
    """
    # Get existing records
    existing_records = get_db_data(cursor)
    
    # Create a set of existing formatted_serial_numbers
    existing_serials = {serial for _, _, serial in existing_records}
    
    # Filter the new data
    mask = ~new_data_df['formatted_serial_number'].isin(existing_serials)

    print(f"Existing records count: {len(existing_records)}")
    print(f"New records count: {len(new_data_df[mask])}")
    print(f"Filtered new records count: {len(new_data_df[~mask])}")
    
    return new_data_df[mask]
```
